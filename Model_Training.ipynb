# Using SAM Model

```python
from ultralytics import YOLO
from ultralytics import SAM

yolomodel = YOLO("/content/runs/detect/train3/weights/best.pt")

results = yolomodel("/content/Te-me_0082_jpg.rf.f2908ba9216d4bbd1ace5352986cb793.jpg")

sammodel = SAM("sam2_b.pt")

for result in results:
  classid = result.boxes.cls.int().tolist()
  if len(classid):
    boxes = result.boxes.xyxy
    sam_results = sammodel(result.orig_img, bboxes = boxes, verbose = False, save = True)
```
